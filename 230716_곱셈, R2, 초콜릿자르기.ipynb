
### 2163.  
- M X N  
- (N-1) + N*(M-1)
- strip()_    
- sys.stdin.readline    
- https://daimhada.tistory.com/180

```python
import sys
input = sys.stdin.readline

def solve(n, m):
    return (n-1) + n * (m-1)

if __name__ == "__main__":
    n, m = map(int, input().strip().split())
    print(solve(n, m))
```
